## Sets in Pandas

- inner join A,B: "A intersect B" (A snitt B)
- outer join A,B: "A union B" (A or B or both)

## Concat

- concatenates pandas objects along an axis (rows=0 or columns=1)
- there are set logics (e.g. inner, outer, left, right)

```python
df1, df2
```

```python
pd.concat([df1, df2]) # axis = 0 is default
```

```python
# all cols from df1 and all cols from df2
# indices missing in one df is filled with NaNs
pd.concat([df1, df2], axis=1)
```

```python
df1, df3
```

```python
# adds df1 rows first, then adds df3 rows
pd.concat([df1, df3])
```

```python
# when joining by columns, intersecting rows are merged
# outer join is default (everything from df1 and df2)
pd.concat([df1, df3], axis = 1)
```

```python
# inner only gives the intersecting index
pd.concat([df1, df3], axis = 1, join = "inner")
```

## Merge

- more complicated than join and concat, but more flexible

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K0", "K1", "K2"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"]
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]
    }
)

left, right
```

```python
# one to many (there are two K0, but only one C0, D0, so they are duplicated)
# inner join default -> K3 from right not included
pd.merge(left, right, on = "key", indicator=True)
```

```python
# syntax with method (equivalent to above)
left.merge(right, on = "key", indicator=True)
```

```python
left.merge(right, on = "key", how = "outer", indicator=True)
```

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "A": ["A0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)

left, right
```

```python
# merge on both key1 and key2 columns, and an inner join (intersection)
# A_x is from left, A_y is from the right
left.merge(right, on = ["key1", "key2"], indicator=True)
```

```python
# merges on left + intersect
left.merge(right, on = ["key1", "key2"], indicator=True, how = "left")
```

```python
left.merge(right, on = ["key1", "key2"], indicator=True, how = "right", suffixes=["_left", "_right"])
```

## Join

- uses merge internally
- combines columns

```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)


right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)

left, right
```

```python
# note we join on K1 which only exists in left -> so default is a left join
left.join(right)
```

```python
# indices from right -> right join
right.join(left)
left.join(right, how = "right")
```
